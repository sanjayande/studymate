```python
def decision_bot(problem):
    problem = problem.lower()

    if "career" in problem or "job" in problem:
        return (
            "This seems like a career decision.\n"
            "Think about:\n"
            "- Long-term growth\n"
            "- Skills you enjoy\n"
            "- Financial stability"
        )

    elif "study" in problem or "exam" in problem:
        return (
            "This is a study-related decision.\n"
            "Consider:\n"
            "- Available time\n"
            "- Your interest level\n"
            "- Pressure vs motivation"
        )

    else:
        return (
            "This is a personal decision.\n"
            "Ask yourself:\n"
            "- What matters most to you?\n"
            "- What will you regret less?"
        )
```

```python
def score_option(option):
    print(f"\nBot: Let's evaluate '{option}'")

    pros = int(input("Bot: How many PROS (0‚Äì5)? "))
    cons = int(input("Bot: How many CONS (0‚Äì5)? "))

    score = pros - cons
    return score
```

```python
def compare_options(option1, option2):
    score1 = score_option(option1)
    score2 = score_option(option2)

    if score1 > score2:
        decision = option1
    elif score2 > score1:
        decision = option2
    else:
        decision = "Both options are equally balanced"

    return score1, score2, decision
```

```python
def confidence_level(score1, score2):
    difference = abs(score1 - score2)
    confidence = min(100, difference * 20)
    return confidence
```

```python
opt1 = input("Bot: Enter Option 1: ")
opt2 = input("Bot: Enter Option 2: ")

s1, s2, final_decision = compare_options(opt1, opt2)
conf = confidence_level(s1, s2)

print("\nBot: Decision Suggestion:", final_decision)
print("Bot: Confidence Level:", conf, "%")
```

```python
def score_option(option):
    print(f"\nü§ñ Bot: Let's evaluate '{option}'")
    pros = int(input("ü§ñ Bot: How many PROS (0‚Äì5)? "))
    cons = int(input("ü§ñ Bot: How many CONS (0‚Äì5)? "))
    return pros - cons


def confidence_level(score1, score2):
    diff = abs(score1 - score2)
    return min(100, diff * 20)


def decision_chatbot():
    print("ü§ñ Bot: Describe your problem briefly.")
    problem = input("You: ")

    print("\nü§ñ Bot: Enter two options.")
    option1 = input("ü§ñ Bot: Enter Option 1: ")
    option2 = input("ü§ñ Bot: Enter Option 2: ")

    score1 = score_option(option1)
    score2 = score_option(option2)

    if score1 > score2:
        decision = option1
    elif score2 > score1:
        decision = option2
    else:
        decision = "Both options are equally balanced"

    confidence = confidence_level(score1, score2)

    print("\nü§ñ Bot: DECISION RESULT")
    print("---------------------")
    print("Problem:", problem)
    print("Suggested Choice:", decision)
    print("Confidence Level:", confidence, "%")
    print("ü§ñ Bot: Final decision is yours üôÇ")


# RUN BOT
decision_chatbot()
3
```

```python
import gradio as gr

def decision_ui(problem, option1, pros1, cons1, option2, pros2, cons2):
    if problem.strip() == "" or option1.strip() == "" or option2.strip() == "":
        return "‚ö†Ô∏è Please fill all fields before submitting."

    score1 = pros1 - cons1
    score2 = pros2 - cons2

    if score1 > score2:
        decision = option1
    elif score2 > score1:
        decision = option2
    else:
        decision = "Both options are equally balanced"

    confidence = min(100, abs(score1 - score2) * 20)

    return f"""
üß† Problem:
{problem}

üîπ Option 1: {option1}
Score: {score1}

üîπ Option 2: {option2}
Score: {score2}

‚úÖ Suggested Choice:
{decision}

üìä Confidence Level:
{confidence} %

‚ö†Ô∏è This is a decision-support system.
"""

gr.Interface(
    fn=decision_ui,
    inputs=[
        gr.Textbox(label="Describe your problem"),
        gr.Textbox(label="Option 1"),
        gr.Slider(0, 5, step=1, label="Pros (Option 1)"),
        gr.Slider(0, 5, step=1, label="Cons (Option 1)"),
        gr.Textbox(label="Option 2"),
        gr.Slider(0, 5, step=1, label="Pros (Option 2)"),
        gr.Slider(0, 5, step=1, label="Cons (Option 2)")
    ],
    outputs=gr.Textbox(label="Decision Result"),
    title="ü§ñ AI Decision-Making Chatbot",
    description="Test by changing sliders and options"
).launch()
```
